```python
import numpy as np
from time import perf_counter
import decimal
import random
import pandas as pd
import matplotlib.pyplot as plt
decimal.getcontext().prec = 100
```

```python
n = np.arange(1, 2001, 1)
j = np.arange(0, 5, 1)
```

```python
def f_1(v):
    return 1
```

```python
def f_2(v):
    i=0
    sum = 0
    while i < len(v):
        sum = sum + v[i]
        i = i + 1
        
    return sum
```

```python
def f_3(v):
    i=0
    sum = 1
    while i < len(v):
        sum = sum * v[i]
        i = i + 1
    return sum
```

```python
def f_naiv(V):
    
    if len(V) > 1:
        result = decimal.Decimal(V[0]) + decimal.Decimal(V[1]*1.5)
        xPower = decimal.Decimal(1.5)
        i = 2
        while i < len(V):
            xPower = decimal.Decimal(xPower)*decimal.Decimal(1.5)
            result = decimal.Decimal(result) + decimal.Decimal(V[i])*decimal.Decimal(xPower)
            i += 1
        return decimal.Decimal(result)
    else:
        return V[0]
```

```python
def f_horners(V):
    n_1 = np.arange(len(V), 0, -1)
    result = decimal.Decimal(0)
    for i in n_1:
        result = decimal.Decimal(result)*decimal.Decimal(1.5)
        result = decimal.Decimal(result) + decimal.Decimal(V[i-1])
    return decimal.Decimal(result)
```

```python
def bSort(V):
    
    
    for i in range(len(V)-1):
        
        for j in range(len(V)-i-1):
            
            if V[j] > V[j+1]:
                
                buff = V[j]
                V[j] = V[j+1]
                V[j+1] = buff
    return V
```

```python
def QuickSort(A):
    if len(A) <= 1:
        return A
    else:
        q = random.choice(A)
        L = []
        M = []
        R = []
        for elem in A:
            if elem < q:
                L.append(elem) 
            elif elem > q: 
                R.append(elem) 
            else: 
                M.append(elem)
        return QuickSort(L) + M + QuickSort(R)
```

```python
minrum = 32
def find_minrun(n): 
    r = 0
    while n >= minrum: 
        r |= n & 1
        n >>= 1
    return n + r
def InsSort(arr,start,end):    
    for i in range(start+1,end+1):
        elem = arr[i]
        j = i-1
        while j>=start and elem<arr[j]:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = elem
    return arr
def merge(arr,start,mid,end):
    if mid==end:
        return arr
    first = arr[start:mid+1]
    last = arr[mid+1:end+1]
    len1 = mid-start+1
    len2 = end-mid
    ind1 = 0
    ind2 = 0
    ind  = start
     
    while ind1<len1 and ind2<len2:
        if first[ind1]<last[ind2]:
            arr[ind] = first[ind1]
            ind1 += 1
        else:
            arr[ind] = last[ind2]
            ind2 += 1
        ind += 1
     
    while ind1<len1:
        arr[ind] = first[ind1]
        ind1 += 1
        ind += 1
              
    while ind2<len2:
        arr[ind] = last[ind2]
        ind2 += 1
        ind += 1   
              
    return arr
            
def TimSort(arr):
    n = len(arr)
    minrun = find_minrun(n)
    for start in range(0,n,minrun):
        end = min(start+minrun-1,n-1)
        arr = InsSort(arr,start,end)
        
    curr_size = minrun
    while curr_size<n:    
        for start in range(0,n,curr_size*2):
            mid = min(n-1,start+curr_size-1)
            end = min(n-1,mid+curr_size)
            arr = merge(arr,start,mid,end)
        curr_size *= 2
    return arr
```

```python
def Matrix(A, B):
    return A.dot(B)
```

```python
time_f_1 = []
for i in n-1: 
    sum = 0
    V = np.random.uniform(0, 1, n[i])
    for k in j:
        t0 = 0
        t1 = 0
        
        t0 = perf_counter()
        f_1(V)
        t1 = perf_counter()
        sum = sum + (t1 - t0)
    time_f_1.append(sum/5)
```
